```python
import pandas as pd
```

```python
df1=pd.read_csv('SEM5.csv')
```

```python
df1.info()
```

```python
grade_points={'O':10,'A+':9,'A':8,'B+':7,'B':6,'C':5,'F':0}
```

```python
df1['DM_pt']=df1['DM'].map(grade_points)*4
df1['FLA_pt']=df1['FLA'].map(grade_points)*3
df1['CN_pt']=df1['CN'].map(grade_points)*4
df1['BD_pt']=df1['BD'].map(grade_points)*3
df1['COMM_pt']=df1['COMM'].map(grade_points)*1
df1['CC_pt']=df1['CC'].map(grade_points)*3
df1['CNS_pt']=df1['CNS'].map(grade_points)*3
df1['FS_pt']=df1['FS'].map(grade_points)*3
df1['SRWC_pt']=df1['SRWC'].map(grade_points)*3
```

```python
columns=['DM_pt','FLA_pt','CN_pt','BD_pt','COMM_pt','CC_pt','CNS_pt','FS_pt','SRWC_pt']
```

```python
df1.fillna(0,inplace=True)
```

```python
total_credits=21
```

```python
df1['SGPA']=df1[columns].sum(axis=1)/total_credits
```
